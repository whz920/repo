来做一个图书管理系统，和之前的ATM很像  
- 查询所有书籍
- 添加书籍
- 借书
- 还书
- 退出系统


我们使用`menu`来封装菜单功能  
整体架构如下:  

```python
class Book:
    def __init__(self, name, author, info, status = 0):
        # 初始化属性信息
        """
           书名，作者，简介，借阅状态
        """
        self.name = name
        self.author = author
        self.info = info
        self.status = status
        pass
    
    def __str__(self):
        if self.status == 0:
            status = '未借阅'
        else:
            status = '已借阅'
        
        return '书名：{}\t 作者：{}\t 简介：{}\t 借阅状态：{}\t'.format(self.name, self.author, self.info, status)
    

if __name__ == 'main':
    book = Book('三体', '刘慈欣', '科幻巨作', 0)
    print(book) 

    


```

```python
class BookManager():
    books = []
    
    def __init__(self):
        book1 = Book('数学之美','吴军','AI的科普读物',0)
        book2 = Book('三国演义','罗贯中','四大名著之一', 1)
        book3 = Book('长安十二时辰','马伯庸','长安侦探小说')
        
        self.books.append(book1)
        self.books.append(book2)
        self.books.append(book3)
    pass
    
    def menu(self):
        #显示菜单
        while True:
            print("""
1、查询
2、捐书
3、借书
4、还书
5、退出
             """)
            choice = int(input('请选择功能对应的数字：'))
            if choice == 1:
                self.show_book()
            elif choice == 2:
                self.add_book()
            elif choice == 3:
                self.borrow_book()
            elif choice == 4:
                return_book()
            elif choice == 5:
                print('感谢使用！')
                break
            else:
                print('ERROR!请重新选择！')
        pass
    
    def add_book(self):
        #捐书
        new_name = input('请输入书籍名称：')
        new_author = input('请输入书籍作者：')
        new_info = input('请输入书籍简介：')
        new_book = Book(new_name, new_author, new_info)
        
        self.books.append(new_book)
        print('书籍已成功保存！')
        pass
    
    def show_book(self):
        #显示所有书
        for book in self.books:
            print(book)
        pass
    
    def borrow_book(self):
        #借书
        borrow_name = input('请输入借阅图书名称：')
        res = self.check_book(borrow_name)
        
        if res :
            if res.status == 1:
                print('你来晚了，{}已经借阅出去了！'.format(borrow_name))
            else:
                res.status = 1
                print('借阅成功！')
        else:
            print('图书馆正在采购ing...')
        pass
    
    def return_book(self):
        #还书
        name = input('请输入归还书籍名称：')
        res = self.check_book(name)
        
        if res :
            if res.status == 1:
                res.status = 0
                print('归还成功，欢迎再次借阅！')
            else:
                print('还未借阅出去！')
        else:
            print('请确认书籍名称！')
        pass
    
    def check_book(self,name):
        for book in self.books:
            if book.name == name:
                return book
            
        return None
    
```

```python
manager = BookManager()
manager.menu()
```
